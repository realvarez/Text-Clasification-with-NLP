# Convertir Modelos keras

```python
yaml_file = open('drive/My Drive/Tesis/RNN/save_1/SINT/modelSINT.yaml', 'r')
loaded_model_yaml = yaml_file.read()
yaml_file.close()
loaded_model = tf.keras.models.model_from_yaml(loaded_model_yaml)
# load weights into new model
loaded_model.load_weights("drive/My Drive/Tesis/RNN/save_1/SINT/modelSINT.h5")
print("Loaded model from disk")
loaded_model.save("drive/My Drive/Tesis/RNN/modelSINT.h5")
print("Saved model")

# converter = tf.lite.TFLiteConverter.from_keras_model_file("modelMA.h5")
# converter = tf.lite.TFLiteConverter.from_keras_model_(loaded_model)
# tflite_model = converter.convert()
```
